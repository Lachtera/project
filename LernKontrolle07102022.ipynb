# Dataframe df gegeben


```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(20, 40))
df.columns = [x for x in 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMN']
df.iloc[3,4] = np.nan
df.iloc[2,0] = np.nan
df.iloc[4,5] = np.nan
df.iloc[0,6] = np.nan
df.head()
```

# Step1
### Ändere die Zahlen in dem gegebenen Dataframe so um, dass diese nur noch 2 Nachkommastellen enthalten.

```python
df.style.format(precision=2)
df.head(5)
```

# Step2
### Gib vom Dataframe df mithilfe von slicing die ersten 5 Zeilen und die ersten 8 Spalten aus und speichere es in df1

```python
df1 = df.loc[:,["a", "b", "c","d", "e", "f", "g", "h"]]
df1.head(5)
```

# Step3
### Färbe die NaN  Werte von df1 in der Tabelle mit Styler Object in der Farbe rot

```python
df1.style.applymap(func=lambda x: "color:red" if pd.isna(x) else None)
```

# Step4
### Färbe die NaN Werte des df1 nur in der Spalte e und g in der Farbe grün mit Styler-Objekt

```python
df1.style.highlight_null("green",subset=["e","g"])
```

# Step5
### Färbe das Minimum jeder Spalte von df1 mit Styler-Objekt.


```python
df1.style.applymap(func=lambda x:"color-green" if pd.isna(x) else None,subset=["e","g"])
```

```python
df1.style.highlight_min(color="red")
```

# Step6
### Färbe das Maximum jeder Spalte von df1 mit styler-Objekt.



```python
df1.style.highlight_max(color="green")
```
